### Slicing Strings
You can return a range of characters by using the slice syntax.

Specify the start index and the end index, separated by a colon, to return a part of the string.

Square brackets can be used to access elements of the string

* Slice From the Start

* Slice To the End

* Negative Indexing

* between

![](https://i1.faceprep.in/Companies-1/string-slicing-in-python.png)

```python
#Slice From the Start

name = "face"
print(name[:2])
```

```python
#slice from the end

print(name[2:])
```

```python
#slice form the negative

print(name[-2:])
```

```python
#slice from the between

print(name[1:3])
```

### Modify Strings

* The upper() method returns the string in **upper case**

* The lower() method returns the string in **lower case**

* **Whitespace** is the space before and/or after the actual text, and very often you want to remove this space.The strip() method removes any whitespace from the beginning or the end

* The replace() method **replaces** a string with another string

* The split() method returns a list where the text between the **specified separator** becomes the list items

```python
#modify the string using upper case

name = "Mew Technology"
print(name.upper())
```

```python
#modify the string using lower case

print(name.lower())
```

```python
#modify the string using whitespace removal - strip

sentence = "     this is my, first day      class    "
print(sentence.strip())
```

```python
#modify the string using replace

print(sentence.replace("first","last"))
```

```python
#modify the string using split

exam ="cat,dog,rat"
print(exam.split(","))
```
